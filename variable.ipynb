### Boolean Variables 

```python
students_status = True
print(type(students_status))
print('Is this student enrolled?', students_status)
students_status = False
print('Is this student enrolled?', students_status)

# Boolean logic

A = True
B = False

print('A and B is', A and B)
print('A or B is', A or B)
print('not A is', not A)
print('not B is', not B)
print('A xor B is', A ^ B)


```

# Fraction Variables

```python
from fractions import Fraction # we import the Fraction class from the fractions module to work with rational numbers
my_fraction = Fraction(1, 3) # we create a fraction representing 1/3
print(my_fraction, type(my_fraction))  # we print the fraction, which will display as 1/3
my_fraction2 = Fraction(2, 3) # we create another fraction representing 2/5
sum_of_fractions = my_fraction + my_fraction2 # we add the two fractions together

print(sum_of_fractions, type(sum_of_fractions)) # we print the result of adding the two fractions, which will be 1/3 + 2/3 = 1

```

# String and their methods 

```python
string_var = "Hello, World!. Today is saturday and the weather is sunny" # we create a string variable containing the text "Hello, World!"
print(string_var, type(string_var)) # we print the string and its type, which will be <class 'str'>

string_var2 = 'ABCDEBFHIJKLMNOPRST.' 
```

```python
str = 'ABCDEBFHIJKLMNOPQRSTUWXYZ'
print(str, type(str)) # we print the second string and its type, which will also be <class 'str'> 
print(str.lower()) # we convert the string to lowercase and print it, which will be "abcdefhijklmnopqrstwxyz"

# slicing string

str_two = 'Python programming is great!' # we create a string variable containing the text "Python programming is great!"
print(str_two[0:6]) # we slice the string to get the first 6 characters, which will be "Python"'
print(str_two[7:18]) # we slice the string to get the characters from index 7 to 17, which will be "programming"
print(str_two[19:21]) # we slice the string to get the characters from index 19 to the end, which will be "is"
print(str_two[22:]) # we slice the string to get the characters from index 22 to the end, which will be "great!" 
print(str_two[:6]) # we slice the string to get the first 6 characters, which will be "Python"

firts_name_and_last_name = "Brayan Roncancio"
print('first name:', firts_name_and_last_name[:6]) # we slice the string to get the first name, which will be "Brayan"

#removing whitespace
string_with_whitespace = "   Hello, World!   " # we create a string variable with leading and trailing whitespace
print(string_with_whitespace) # we print the string with whitespace, which will display as "   Hello, World!   "
print(string_with_whitespace.strip()) # we remove the leading and trailing whitespace and print the result, which will be "Hello, World!"

#string concatenation
first_name = "Brayan"
last_name = "Roncancio"
full_name = first_name + " " + last_name # we concatenate the first name and last name with a space in between
print('Full name:', full_name) # we print the full name, which will be "Brayan Roncancio"
```

### F-String: Very piwerfull way for formating strings


```python
age = 26
# This will produce an error:
txt = ' My name is Brayan, I am' + str(age) # we concatenate a string with the age variable, converting the age to a string first
print(txt) # we print the resulting string, which will be " My name is Brayan, I am 26 years old."
```

```python
name = 'Brayan'
age = 26 
txt = 'My name is {} and I am {} years old.'.format(name, age) # we use the format method to insert the name and age into the string
print(txt) # we print the formatted string, which will be "My name is Brayan and I am 26 years old."
```
