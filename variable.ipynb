### Boolean Variables 

```python
students_status = True
print(type(students_status))
print('Is this student enrolled?', students_status)
students_status = False
print('Is this student enrolled?', students_status)

# Boolean logic

A = True
B = False

print('A and B is', A and B)
print('A or B is', A or B)
print('not A is', not A)
print('not B is', not B)
print('A xor B is', A ^ B)


```
