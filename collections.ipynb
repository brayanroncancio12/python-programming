## Data collections

In Python, composite data types (also known as compound data type collection) are containers that hold multiple values, which can include other composite types. The main built-in composite data types are list, tuples, dictonaries, and sets






```python
my_list = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]    # Create a list of numbers from 0 to 9 , we create a list with brackets and separate the items with commas
print(type(my_list))    # Print the type of the list, we use the type() function to check the type of the variable
```

```python
my_list #print the whole lists
```

```python
my_list[0] # Access the first element of the list 
my_list[3] # Access the fourth element of the list
my_list[-1] # Access the last element of the list using negative indexing
my_list[0:5] # Access the first five elements of the list using slicing
```

```python
my_list[ :3] # Access the first three elements of the list using slicing
my_list[3: ] # Access all elements of the list starting from the fourth element using slicing
```

```python
# for loops and lists
my_list = [ 10, 20, 30, 40, 50 ,60]
for item in my_list: # Use a for loop to iterate through the elements of the list
    item = item + 5 # Print each element of the list with 5 added to it
    print(item)
```

```python
my_list = [ 10, 20, 30, 40, 50 ,60]
variable = range(len(my_list))
print(variable) # Use the range() function to create a sequence of numbers from 0 to the length of the list
```

```python
# adding the 5 to each element of a list using a for loop
my_list = [ 10, 20, 30, 40, 50 ,60]
for i in range(len(my_list)): # Use a for loop with the range() function to iterate through the indices of the list
    my_list[i] += 5 # Add 5 to each element of the list using its index
    print(my_list) # Print each element of the list with 5 added to it
```

```python
my_dict = {'name': 'brayan', 'age': 30, 'city': 'new york'} # Create a dictionary with key-value pairs
print(type(my_dict)) # Print the type of the dictionary, we use the type() function to check the type of the variable
print(my_dict) # Print the whole dictionary
```

```python
print(my_dict['name']) # Access the value associated with the key 'name' in the dictionary
print(my_dict['age']) # Access the value associated with the key 'age' in the dictionary
print(my_dict['city']) # Access the value associated with the key 'city' in the dictionary
```
