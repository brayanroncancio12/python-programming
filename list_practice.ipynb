### Practice 1A

Create a list named 'my_list' with these integers:
[10, 20, 30, 40, 50, 60]
Then:
- Print its type.
- Print the first element.
- Print the fourth element.
- Print the last element using negative indexing.


```python
my_list = [10, 20, 30, 40, 50, 60 ]
print(type(my_list))
print(my_list[0])
print(my_list[3])
print(my_list[-1])
```

### Practice 1B (Slicing)

Using the same my_list, print:

- The sub-list containing 30, 40, 50
- Everything from the start up to 40 (inclusive)
- Everything from 30 to the end

```python
print(my_list[2:5])
print(my_list[:4])
print(my_list[2:])
```
