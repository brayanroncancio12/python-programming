-2.4 Name Cases Use a variable to represent a person's name, and then print that person's name in lowcase, uppercase, and litle case. 

```python
name = input('Enter your name: ')

if name != None:
    print (name.upper())
    print (name.lower())
    print (name.title())
else:
    print('No name provided.')
```
