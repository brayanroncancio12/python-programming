# Mathematical Operations In Python

```python
# Addition 
c = 10 + 45
print("The sum of and 45 is:", c)
```

```python
# Subtraction
d = 100 - 340
print("The difference between 100 and 340 is:", d)
```

```python
# Multiplication
e = 25 * 4
print("The product of 25 and 4 is:", e)
```

```python
# Division
#float division
f = 100 / 3
print("The quotient of 100 divided by 5 is:", f)
```

```python
#Division
#floatint division
g = 100 // 3
print("The quotient of 100 divided by 5 is:", g)
```

```python
# exponentiation

g = 2 ** 3
print("2 raised to the power of 3 is:", g)  
```

```python
import math #importing the math module to use mathematical functions

#return remainder
print(math.fmod(9, 2)) 
print(math.fmod(9, 3)) 
print(math.fmod(18, 4)) 
```

```python
#square root
print(math.sqrt(16)) #square root of 16 is 4
print(math.sqrt(25)) #square root of 25 is 5
print(math.sqrt(36))#square root of 36 is 6
```

```python
# Convert long lines multiple lines by \

value_1 = 10
value_2 = 20
value_3 = 30
value_4 = 40
value_5 = 50
value_6 = 60

# long line 
result = value_1 + value_2 + value_3 + value_4 + value_5 + value_6
print("The sum of the values is:", result) 

#short line
result = value_1 + value_2 + \
        value_3 +  value_4 + \
        value_5 + value_6

print("The sum of the values is:", result)
```

```python
### Bad code with magic numbers
laptop_price = 1000
final_price = laptop_price - (laptop_price * 0.10)* 1.13 # adding 15% tax

print("The final price of the laptop is:", final_price)

### Good code without magic numbers
laptop_price1 = 1000
discount_percentage1 = 0.10
tax_percentage1 = 1.13
final_price1 = (laptop_price1 - (laptop_price1 * discount_percentage1))*tax_percentage1
print("The final price of the laptop is:", final_price)
```
